```python
import matplotlib.pyplot as plt
from sklearn.datasets.samples_generator import make_blobs
import pandas as pd
import numpy as np
from sklearn import svm
from mlxtend.plotting import plot_decision_regions
from sklearn.linear_model import LogisticRegression
```

```python
clf = svm.LinearSVC()
```

```python
def plotDecisionBoundaryLinear(weight, bias,randomVar):
    X, y = make_blobs(n_samples=100, centers=2, n_features=2,random_state=randomVar)
    df = pd.DataFrame(dict(x=X[:,0], y=X[:,1], label=y))
    colors = {0:'red', 1:'blue'}
    fig, ax = plt.subplots()
    grouped = df.groupby('label')
    for key, group in grouped:
        group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, color=colors[key])
    x_space = np.linspace(-10,10,100)
    y_space =-((weight[0]/weight[1])*x_space) -(bias/weight[1])
    plt.plot(x_space, y_space, '-', label='Test')
    plt.show()
    
```

```python
def plotDecisionBoundaryNonLinear(weight, bias,NonLinear_type):
    if(NonLinear_type=='moon'):
        X,y= make_moons(150,2,random_state=2)
    else:
        X,y= make_circles(150,2,random_state=2)
    df = pd.DataFrame(dict(x=X[:,0], y=X[:,1], label=y))
    colors = {0:'red', 1:'blue'}
    fig, ax = plt.subplots()
    grouped = df.groupby('label')
    for key, group in grouped:
        group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, color=colors[key])
    if(NonLinear_type=='moon'):
        x_space = np.linspace(-1,2,10)
    else:
        x_space = np.linspace(-1,1,10)
    y_space =-((weight[0]/weight[1])*x_space) -(bias/weight[1])
    plt.plot(x_space, y_space, '-', label='Test') 
    plt.show()
```

```python
def MakeDataset(randomVar):
    X, y = make_blobs(n_samples=100, centers=2, n_features=2,random_state=randomVar)
    df = pd.DataFrame(dict(x=X[:,0], y=X[:,1], label=y))
    colors = {0:'red', 1:'blue'}
    fig, ax = plt.subplots()
    grouped = df.groupby('label')
    for key, group in grouped:
        group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, color=colors[key])
    plt.show()
    return X,y
```

```python
def perceptronAlgo(X,y,weight,typeLine='NonLinear',randomVar=2,NonLinear_type='circle'):
    bias=0
    count=0
    didItChange=True
    while didItChange is True and count<15:
        didItChange=False
        count=count+1
        #print("Weight is obtained as ",weight)
        for i in range(X.shape[0]):
            x=X[i]
            #x=np.expand_dims(x, axis=0)
            t=y[i]
            yDash=bias+(np.dot(weight,x))
            #print(yDash)
            #print(yDash)
            if(yDash>0):
                yDash=1.0
            else:
                yDash=-1.0
            #print(yDash,"  ",t)
            if yDash != t:
                #print("Comes here for ",yDash,"  ",t)
                didItChange=True
                weight=weight+learning_rate*x*t
                bias=bias+learning_rate*t
        if typeLine == "Linear":
            plotDecisionBoundaryLinear(weight, bias,randomVar)
        else:
            print("Weight : ", weight," Bias: ",bias)
            plotDecisionBoundaryNonLinear(weight, bias,NonLinear_type)
    return weight, bias
    #print(count)
```

## Linear Dataset-1

```python
X,y = MakeDataset(10)
y[y==0]=-1
```

```python
# definition of the variables that are needed
```

```python
learning_rate=0.01
yDash=0
```

### Algorithm with weight as [0,0]

```python
weight, bias = perceptronAlgo(X,y,[0,0],'Linear',10)
```

### Algorithm with weight as [-5,-5]

```python
weight, bias = perceptronAlgo(X,y,[-5,-5],'Linear',10)
```

```python
y[y==-1]=0
```

### SVM Implementation for Dataset 1

```python
clf.fit(X, y)
plot_decision_regions(X=X,y=y,clf=clf, legend=2)
plt.title('SVM Decision Region Boundary', size=14)
plt.show()
```

### Logistic Regression Implementation for Dataset 1

```python
clf1= LogisticRegression(random_state=0, solver='lbfgs',multi_class='multinomial').fit(X, y)
plot_decision_regions(X=X,y=y,clf=clf1, legend=2)
plt.title('Logistic  Decision Region Boundary', size=14)
plt.show()
```
