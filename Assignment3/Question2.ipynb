### This is the numerical implementation of the gradient descent algorithm. 
<br>
Here ,<br>
the gradient of the error wrt the weight w1 is ::: lambda1 * w1 <br>
the gradient of the error wrt the weight w2 is ::: lambda2 * w2 <br>

I assumed the weights  as w1=1;w2=1.<br>
For various learning rates of [0.1,0.01,0.001,0.0001] and for various values of lambda1 and lambda2 the graph for error value vs epochs was plotted.

### Observation:
It was observed that the error kept reducing to 0 for small values of w1,w2. If the w1 or w2 were given large values or if the learning ates were greater than or equal to 1, we observed that error shot up exponentially. This was mainly because of the gradient calculation(The above example is not shown here in the code)
<br>
Normally, from the graphs, we observe that bigger learning rates converge very fast compared to the smaller learning rates because of the multiplication( smaller steps towards the minimum value). Also, it depended on the lambda1,lambda2 valuess(i.e on their ratios).<br>
We can see that, the learning rate of 0.1 which converges faster in the examples shown, still depends on the ratio of lambda1 and lambda2. This is shown very clearly in the graphs 3(lambda1=0.1,lambda2=1) and 4(lambda1=1,lambda2=1) . In this particular example, ratios are 1:10 and 1:1. Here we can see that  the 1:1(graph3) converges a little faster than the 1:10(graph4). It can also be seen in graph1  and 2 (but not that clearly because of the large values of lambda1 and lambda2) <br>
Even  1:15 converges faster than the 1:5 (last 2 graphs)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def grad_W1(lambda1, w1):
    return lambda1*w1
```

```python
def grad_W2(lambda2, w2):
    return lambda2*w2
```

```python
def plotError(lambda1, lambda2):
    epochs=1000
    for learning_rate in [0.1,0.01,0.001,0.0001]:
        errors=[]
        w1= 1
        w2= 1
        for  i in range(epochs):
            error = 0.5 * (lambda1*w1*w1 + lambda2*w2*w2)
            errors.append(error)
            w1 = w1 - learning_rate * grad_W1(lambda1, w1)
            w2 = w2 - learning_rate * grad_W2(lambda2, w2)
        plt.plot(errors)
    plt.legend(['0.1','0.01','0.001','0.0001'],loc='upper right')
    plt.title('error value vs epochs')
    plt.show()
```

```python
plotError(1,10)
```

```python
plotError(10,10)
```

```python
plotError(0.1,1)
```

```python
plotError(1,1)
```

```python
plotError(1,5) #ratio : 1:5
```

```python
plotError(1,15) # ratio : 1:15 . converges faster than 1:5 
```
