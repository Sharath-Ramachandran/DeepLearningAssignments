```python
lambda1 = 1
lambda2 = 10
learning_rate = 0.1
epochs =1000
w1 = 1
w2 = 1
```

```python
epochs=100
for learning_rate in [0.1,0.01,0.001,0.0001]:
    errors=[]
    w1= 1
    w2= 1
    for  i in range(epochs):
        errors.append(0.5 * (lambda1*w1*w1 +lambda2*w2*w2))
        w1 = w1 -learning_rate * grad_W1()
        w2 = w2 -learning_rate * grad_W2()
        #print(w1,"   ",w2," ",0.5 * (lambda1*w1*w1 +lambda2*w2*w2))
    plt.plot(errors)
    #print(errors)
plt.legend(['0.1','0.01','0.001','0.0001'],loc='upper right')
plt.show()
```
